```python
diccionario_ascii = {
    'a': 97,  'b': 98,  'c': 99,  'd': 100, 'e': 101,
    'f': 102, 'g': 103, 'h': 104, 'i': 105, 'j': 106,
    'k': 107, 'l': 108, 'm': 109, 'n': 110, 'o': 111,
    'p': 112, 'q': 113, 'r': 114, 's': 115, 't': 116,
    'u': 117, 'v': 118, 'w': 119, 'x': 120, 'y': 121,
    'z': 122
}



def cifrado_cesar(texto, desplazamiento):
    resultado = ""

    for char in texto:
        if 'a' <= char <= 'z':
            # convertir a ASCII
            ascii_val = ord(char) - 97

            # aplicar desplazamiento con mod 26
            nuevo_val = (ascii_val + desplazamiento) % 26

            # volver a caracter
            resultado += chr(nuevo_val + 97)
        else:
            # dejar igual si no es letra
            resultado += char

    return resultado



def decifrado_cesar(texto, desplazamiento):
    resultado = ""

    for char in texto:
        if 'a' <= char <= 'z':
            # convertir a ASCII
            ascii_val = ord(char) - 97

            # aplicar desplazamiento con mod 26
            nuevo_val = (ascii_val - desplazamiento) % 26

            # volver a caracter
            resultado += chr(nuevo_val + 97)
        else:
            # dejar igual si no es letra
            resultado += char

    return resultado

mensaje = "hola mundo"
print("mensaje original: ", mensaje)
print("mensaje cifrado: ",cifrado_cesar(mensaje, 3))

mensaje = cifrado_cesar(mensaje, 3)
print("mensaje de decifrado",decifrado_cesar(mensaje, 3))





```

```python
def vigenere_encrypt(plaintext, key):
    plaintext = plaintext.upper()
    key = key.upper()
    ciphertext = ""
    key_index = 0

    for char in plaintext:
        if char.isalpha():
            p = ord(char) - ord('A')
            k = ord(key[key_index % len(key)]) - ord('A')
            c = (p + k) % 26
            ciphertext += chr(c + ord('A'))
            key_index += 1
        else:
            ciphertext += char  # keep spaces/punctuation

    return ciphertext




def vigenere_decrypt(ciphertext, key):
    ciphertext = ciphertext.upper()
    key = key.upper()
    plaintext = ""
    key_index = 0

    for char in ciphertext:
        if char.isalpha():
            c = ord(char) - ord('A')
            k = ord(key[key_index % len(key)]) - ord('A')
            p = (c - k) % 26
            plaintext += chr(p + ord('A'))
            key_index += 1
        else:
            plaintext += char

    return plaintext
```

```python
message = "hola"
key = "sol"

encrypted = vigenere_encrypt(message, key)
print("Encrypted:", encrypted)

decrypted = vigenere_decrypt(encrypted, key)
print("Decrypted:", decrypted)
```
