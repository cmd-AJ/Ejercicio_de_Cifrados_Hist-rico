```python


def cifrado_cesar(texto, desplazamiento):
    resultado = ""

    for char in texto:
        if 'a' <= char <= 'z': #es abecedario entonces tiene que estar de la a hasta la z
            # convertir a ASCII
            ascii_val = ord(char) - 97

            # aplicar desplazamiento con mod 26 la formula aplicada es ( x + desplazamiento ) % mod 26
            nuevo_val = (ascii_val + desplazamiento) % 26


            # retorna el caracter a caracter en ascii que empieza del 97 "a"
            resultado += chr(nuevo_val + 97)
        else:
            # dejar igual si no es letra
            resultado += char

    return resultado




# Se realiza lo mismo solo que en la formula no se deplaza hacia adelante sino que atras
def decifrado_cesar(texto, desplazamiento):
    resultado = ""

    for char in texto:
        if 'a' <= char <= 'z':
            # convertir a ASCII
            ascii_val = ord(char) - 97

            # aplicar desplazamiento con mod 26 la formula solo resta su desplazamiento para desencriptar
            nuevo_val = (ascii_val - desplazamiento) % 26

            # volver a caracter
            resultado += chr(nuevo_val + 97)
        else:
            # dejar igual si no es letra
            resultado += char

    return resultado

mensaje = input("que mensaje deseas enviar")
print("mensaje original: ", mensaje)
print("mensaje cifrado: ",cifrado_cesar(mensaje, 3))

mensaje = cifrado_cesar(mensaje, 3)
print("mensaje de decifrado",decifrado_cesar(mensaje, 3))





```

```python


#sin usar nada más solo metemos el cifrado de cesar y determinar el desplazamiento fijo que es de 13 solo. 
def rot13cifrado(texto):

    temporal = cifrado_cesar(texto, 13)

    return temporal

#Desencriptar es lo mismo que rot13 cifrado solo que ponemos -13 como desplazamiento a la izquierda
def rot13decifrado(texto):

    temporal = cifrado_cesar(texto, -13)

    return temporal

mensaje = "hola mundo"
print("mensaje original: ", mensaje)
print("mensaje cifrado: ",rot13cifrado(mensaje))

mensaje = rot13cifrado(mensaje)
print("mensaje de decifrado",rot13decifrado(mensaje))





```

```python
def cifrado_devigenere(plaintext, llave):
    plaintext = plaintext.upper()
    llave = llave.upper()
    mensaje = ""
    llave_index = 0

    for char in plaintext:
        if char.isalpha():
            p = ord(char) - 65
            k = ord(llave[llave_index % len(llave)]) - 65
            c = (p + k) % 26
            mensaje += chr(c + 65)    #En este metodo solo se suma los valores para luego tener el valor del caracter
            llave_index += 1
        else:
            mensaje += char  # keep spaces/punctuation

    return mensaje




def decifrado_devigenere(mensaje, llave):
    mensaje = mensaje.upper() #Convertimos el mensaje en mayusculas
    llave = llave.upper()
    mensajetemporal = ""
    llave_index = 0

    for char in mensaje:
        if char.isalpha():
            c = ord(char) - 65
            k = ord(llave[llave_index % len(llave)]) - 65
            p = (c - k) % 26
            mensajetemporal += chr(p + 65)  #En este metodo solo se suma los valores para luego tener el valor del caracter
            llave_index += 1
        else:
            mensajetemporal += char

    return mensajetemporal
```

```python
message = "hola"
llave = "sol"

encrypted = cifrado_devigenere(message, llave)
print("Cifrado:", encrypted)

decrypted = decifrado_devigenere(encrypted, llave)
print("Decifrado:", decrypted)
```

```python
def analisis_frecuencias(texto):
    texto = texto.upper()   # solo se ponen en mayusculas 
    
    frecuencias = {}
    total = 0

    # Contar letras de A a la Z
    for c in texto:
        if 'A' <= c <= 'Z':
            total += 1
            if c in frecuencias:
                frecuencias[c] += 1
            else:
                frecuencias[c] = 1

# Se genera la tabla de frecuencia dependiendo de que tan comun es la letra 
    print("Análisis de frecuencias:\n")
    print(f"{'Letra':<6} {'Cantidad':<10} {'Frecuencia (%)'}")
    print("-" * 30)

    # Ordenar por frecuencia (de mayor a menor)
    for letra in sorted(frecuencias, key=frecuencias.get, reverse=True):
        cantidad = frecuencias[letra]
        frecuencia = (cantidad / total) * 100  #Para hacer el porcentaje de cada uno se realiza la siguiente formula osea por cada letra guardada tiene su valor dividio el total de la cadena de caracteres
        print(f"{letra:<6} {cantidad:<10} {frecuencia:.2f}")

    return frecuencias



texto_cifrado = "KHOOR ZRUOG"
analisis_frecuencias(texto_cifrado)
```
